```python

f = open('/project_data/data_asset/train_label.csv')
#label_list = f.readlines()
label_list = f.read().splitlines()
f.close()
```

```python
label_list
```

```python
train_list = []
dir_list = []
with open('/project_data/data_asset/train_img.csv') as f2:
    for line in f2:
        line_split = line.strip().split("_")
        image_number = line_split[0]
        img_dir = line_split[1]
        train_list.append(image_number)
        dir_list.append(img_dir)
        print(image_number)
        print(img_dir)
```

```python
import pandas as pd
df = pd.DataFrame({
    'image_num':train_list,
    'img_dir':dir_list,
    'label':label_list
    
})
```

```python
df['image_num'] = df['image_num'].astype(int)
df['label'] = df['label'].astype(int)
```
